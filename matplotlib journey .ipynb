```python
a=np.arange(0,20,2)
from matplotlib import style
b=np.arange(20,40,2)
style.use("ggplot")
plt.grid(visible=None, which='minor', axis='both',linestyle='-')
plt.plot(a,b,color='r',marker='s',linestyle="--")
plt.title("MAYA CHECK ",fontsize=24)
plt.xlabel("TIMRO MAYA",fontsize=17)
plt.ylabel("MERO MAYA",color='b',fontsize=17)
plt.legend(["temp line "],loc=4)
plt.show()
```

```python
a=np.arange(0,20,2)
from matplotlib import style
b=np.arange(20,40,2)
c=np.arange(30,50,2)
style.use("ggplot")
plt.grid(visible=None, which='minor', axis='both',linestyle='-')
plt.plot(a,b,color='r',marker='s',linestyle="--",label="mero ")
plt.plot(a,c,color='b',marker='s',linestyle="--",label="timro ")
plt.title("MAYA CHECK ",fontsize=24)
plt.xlabel("TIMRO MAYA",fontsize=17)
plt.ylabel("MERO MAYA",color='b',fontsize=17)
plt.legend(loc=4)
plt.show()
```

# histogram 

```python
import matplotlib.pyplot as plt
import numpy as np 
import random 
```

```python
ml_student_age=np.random.randint(18,45,(100))
py_student_age=np.random.randint(15,40,(100))
```

```python
plt.hist(ml_student_age,rwidth=.7,label="machine learning ")
plt.hist(py_student_age,rwidth=0.7,label="python ",)
plt.title("machine learning student ")
plt.xlabel("student")
plt.legend(loc=4)
plt.ylabel("age")
```

# ploting Bar chart

```python
classes=["python","R","AI",'HTML',"DS"]
class1=[30,20,25,10,24]
class2=[40,50,5,20,10]
class3=[35,5,30,15,20]
```

```python
plt.bar(classes,class1,label="class1",width=0.2,align="edge")
plt.bar(classes,class2,label="class2")
style.use('ggplot')
plt.figure(figsize=(16,9))
plt.legend()
```

```python
plt.bar(classes,class1,label="class1",width=0.2,align="edge")
plt.bar(classes,class2,label="class2")
style.use('ggplot')
plt.xlabel("classes")

plt.legend()
```

```python

clas=np.arange(len(classes))
width=0.2
plt.bar(clas,class1,label="class1")
plt.bar(clas+width,class2,label="class2")
style.use('ggplot')
plt.xlabel("classes")
plt.xticks(clas+width,classes,rotation=20)

plt.legend()
```
